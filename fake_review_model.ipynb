```python
cleaned_fake_reviews = pd.read_csv("data/cleaned_fake_reviews.csv")
```

```python
cleaned_fake_reviews
```

```python
cleaned_fake_reviews = cleaned_fake_reviews.drop("Unnamed: 0", axis=1)
```

```python
cleaned_fake_reviews.head()
```

```python
cleaned_fake_reviews["label"].value_counts()
```

```python
cleaned_fake_reviews["is_fake"] = np.where(cleaned_fake_reviews["label"], 0, 1)
```

```python
cleaned_fake_reviews["is_fake"].value_counts()
```

```python
import re

def clean_text(text):
    text = text.lower() # Converts all characters to lowercase
    text = re.sub(r"[^a-z\s]", "", text) # remove numbers, emojis, keeps only letters(a-z) and spaces
    return text

cleaned_fake_reviews["review_text"] = cleaned_fake_reviews["review_text"].apply(clean_text)
```

```python
cleaned_fake_reviews.sample(10)
```

```python
# Importing TF-IDF for text extraction
from sklearn.feature_extraction.text import TfidfVectorizer

# Keep only the top 5000 most important words, stop_words Automatically removes common words like: the, is, was, and, to
tfidf = TfidfVectorizer(max_features = 5000, stop_words="english")

X = tfidf.fit_transform(cleaned_fake_reviews["review_text"]) # Text transformed into numbers using TF-IDF formulas
y = cleaned_fake_reviews["is_fake"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
np.random.seed(42)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

model.score(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score=cross_val_score(model, X, y, cv=10)
cross_val_score
```

```python
np.mean(cross_val_score)
```

```python
# Cross-Validated accuracy
print(f"The cross-validated accuracy is: {np.mean(cross_val_score)*100:.2f}%")
```

```python
fake_test_reviews = [
    "Best company ever!!! Amazing!!! Highly recommended!!!",
    "The work environment is supportive, management is approachable, and there are good learning opportunities for freshers",
    "Awesome awesome awesome awesome awesome Awesome awesome awesome awesome awesome",
    "Very happy with this product. The quality feels solid and delivery was on time.", 
    "I have been using this for two weeks now and it works exactly as described.",
    "The product stopped working after a few days which was disappointing.",
    "Quality is average for the price and customer support was slow to respond.",
    "This product is very good and very useful for daily needs and purposes.",
    "The product quality is good and the product works as expected.", 
    "Absolutely perfect product best ever must buy without thinking totally life changing.",
    "Five stars amazing excellent outstanding value highly recommended to everyone."
]
X_test_fake = tfidf.transform(fake_test_reviews)
fake_preds = model.predict(X_test_fake)

for review, pred in zip(fake_test_reviews, fake_preds):
    label = "Fake" if pred == 1 else "Genuine"
    print(f"{label} : {review}")
```

```python
fake_probs = model.predict_proba(X_test_fake)[:, 1]
threshold = 0.45
fake_preds = (fake_probs >= threshold).astype(int)
```

```python
for review, prob, pred in zip(fake_test_reviews, fake_probs, fake_preds):
    label = "Fake" if pred == 1 else "Genuine"
    print(f"{label} | {prob:.2f} | {review}")
```
